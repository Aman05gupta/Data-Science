# Outlier Removal Using IQR

```python
q1 = dataset6["loan_amnt"].quantile(0.25)
q3 = dataset6["loan_amnt"].quantile(0.75)
```

```python
q1
```

```python
q3
```

```python
IQR = q3 - q1
```

```python
min_range = q1 - (1.5 * IQR)
max_range = q3 + (1.5 * IQR)
```

```python
min_range, max_range
```

```python
new_dataset6 = dataset6[dataset6["loan_amnt"]<=max_range]
```

```python
new_dataset6.shape
```

```python
sns.boxplot(x = "loan_amnt",data=new_dataset6)
plt.show()
```

# Outlier Removal Using Distdistribution

```python
sns.distplot(dataset6["loan_amnt"])
plt.show()
```

```python
min_range = dataset6["loan_amnt"].mean() - (3*dataset6["loan_amnt"].std())
max_range = dataset6["loan_amnt"].mean() + (3*dataset6["loan_amnt"].std())
```

```python
min_range,max_range
```

```python
new_data6 = dataset6[dataset6["loan_amnt"]<=max_range]
```

```python
sns.boxplot(x = "loan_amnt",data=new_data6)
plt.show()
```

```python
z_score = (dataset6["loan_amnt"] - dataset6["loan_amnt"].mean())/(dataset6["loan_amnt"].std())
```

```python
z_score
```

```python
z_score > 3
```

```python
dataset6["z_score"] = z_score
```

```python
dataset6.head(3)
```

# Function Transformer

```python
dataset11 = pd.read_csv("lending_club_loan_two.csv")
dataset11.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.distplot(dataset11["loan_amnt"])
plt.show()
```

```python
q1 = dataset11["loan_amnt"].quantile(0.25)
q3 = dataset11["loan_amnt"].quantile(0.75)
iqr = q3 - q1
```

```python
min_range = q1 - (1.5 * iqr)
max_range = q3 + (1.5 * iqr)
min_range, max_range
```

```python
dataset11[dataset11["loan_amnt"]<=max_range]
```

```python
sns.distplot(dataset11["loan_amnt"])
plt.show()
```

```python
import numpy as np
from sklearn.preprocessing import FunctionTransformer
```

```python
ft = FunctionTransformer(func = np.log1p)
```

```python
ft.fit(dataset11[["loan_amnt"]])
```

```python
dataset11["loan_amnt_tf"] = ft.transform(dataset11[['loan_amnt']])
```

```python
plt.subplot(1,2,1)
plt.title("Befor")
sns.distplot(dataset11["loan_amnt"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset11["loan_amnt_tf"])
plt.show()
```

```python
ft1 = FunctionTransformer(func = lambda x:x**2)
ft1.fit(dataset11[['loan_amnt']])
dataset11['loan_amnt_tf1'] = ft1.transform(dataset11[['loan_amnt']])
```

```python
plt.subplot(1,2,1)
plt.title("Befor")
sns.distplot(dataset11["loan_amnt"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset11["loan_amnt_tf1"])
plt.show()
```
